#### Dictionary 

```python
d={"Yes":1,"No":0}
#Yes-No / 1-0 değiştirdim
```

```python
df["Attrition"]=df["Attrition"].map(d)
#eski değerin yeni değer map.
#map for dögüsünün modern versiyonu
```

```python
df["Attrition"]
```

```python
bos=df[df["Attrition"].isnull()]
```

```python
dolu=df[df["Attrition"].notnull()]
```

```python
bos.shape
```

```python
dolu.shape
```

```python
x=dolu.drop("Attrition",axis=1)
# x tanımladık,Attrition sil kalanları bizim olsun dedik 
y=dolu["Attrition"]
```

```python
#axis=1 :sütun silmek için 
#axis=0 :satır silmek için
```

```python
x
```

```python
x["BusinessTravel"].value_counts() #İş seyahatleri
```

```python
#bunları rakama çeviriceğiz
x=pd.get_dummies(x,drop_first=True)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
g=GaussianNB()
b=BernoulliNB()
```

```python
g.fit(x,y)
b.fit(x,y)
```

```python
tahmin1=g.predict(x)
tahmin2=b.predict(x)
```

```python
# Classificationda başarı oranımız için Accuracy, Precision, Recall, F1 score ve confusion matrix bakmalıyız
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
accuracy_score(tahmin1,y)
```

```python
confusion_matrix(tahmin1,y)
```

```python
print(classification_report(tahmin1,y))
```

```python
sns.heatmap(confusion_matrix(tahmin1,y),annot=True)
```

```python
accuracy_score(tahmin2,y)
```
